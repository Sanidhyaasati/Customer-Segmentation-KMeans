```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df['Age'] = 2026 - df['Year_Birth']
columns_to_drop = ['ID', 'Year_Birth', 'day', 'month']

df = df.drop(columns = columns_to_drop)
```

```python
numeric_cols = df.drop(columns = ['Education', 'Marital_Status']).columns.tolist()
nominal_cols = ['Marital_Status']
ordinal_cols = ['Education']

categories = [['Basic', '2n Cycle', 'Graduation', 'Master', 'PhD']]
ordinal_pipeline = Pipeline(steps=[
    ('encode', OrdinalEncoder(categories=categories)),
    ('scale', StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ('trf1', StandardScaler(), numeric_cols),
    ('trf2', OneHotEncoder(drop = 'first', sparse_output = False), nominal_cols),
    ('trf3', ordinal_pipeline, ordinal_cols)
],remainder = 'passthrough')

preprocessor.set_output(transform="pandas")
df_scaled = preprocessor.fit_transform(df)

df_scaled.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

wcss = [] # This list will hold the Inertia score for each K

# Test K from 1 to 10
for i in range(1, 11):
    # random_state=42 ensures you get the exact same results every time you run it
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(1, 11), y=wcss, marker='o')
plt.title('The Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.xticks(range(1, 11))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
kmeans_final = KMeans(n_clusters=3, init='k-means++', random_state=42)
df['Customer_Segment'] = kmeans_final.fit_predict(df_scaled)
columns_to_profile = ['Income', 'Total_spend', 'Total_kids', 'Age']
cluster_profile = df.groupby('Customer_Segment')[columns_to_profile].mean().round(1)

print(cluster_profile)
```

```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(df_scaled)

plt.figure(figsize=(10, 8))

sns.scatterplot(
    x=tsne_results[:, 0], 
    y=tsne_results[:, 1], 
    hue=df['Customer_Segment'], 
    palette='viridis',
    s=60,
    alpha=0.8
)

plt.title('Customer Segments in 2D Space (t-SNE)', fontsize=16)
plt.legend(title='Customer Segment')
plt.show()
```
